```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
#CONVERTIR BASE DE DATOS A INFORMACIÓN EXTRAIBLE
df= pd.read_excel("Detalle precios y productos fabricados 2022.xlsx")
df.info()
```

**APLICACIÓN DE LOS 10 FILTROS**

```python
#FILTRO POR OBJETO
filtro1= df[(df["NOMBRE_VENDEDOR"]) == "LETICIA RAMIREZ HERNANDEZ"]
filtro1
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV (lo crea :v)
filtro1.to_csv("Filtro1.csv")
```

```python
#FILTRO POR FILAS
filtro2= df.iloc[3:10,: ]     #filas consecutivas
filtro2
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV (lo crea :v)
filtro2.to_csv("Filtro2.csv")
```

```python
#FILTRO POR COLUMNAS 
filtro3= df.iloc[: , 1:3]   #columnas consecutivas
filtro3
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV (lo crea :v)
filtro3.to_csv("Filtro3.csv")
```

```python
#CREACION DE UN INDEX
df1= pd.read_excel("Detalle precios y productos fabricados 2022.xlsx", index_col=1)
df1
```

```python
#FILTRO FILAS VS COLUMNAS
filtro4= df1.loc[["2022-01-05", "2022-12-23"],["PRECIO_UNITARIO"]]
filtro4
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV (lo crea :v)
filtro4.to_csv("Filtro4.csv")
```

```python
#FILTRO POR CABECERA (primeros 5 datos)
filtro5= df.head()
filtro5
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV (lo crea :v)
filtro5.to_csv("Filtro5.csv")
```

```python
#FILTRO COLA (ultimos 5 datos)
filtro6= df.tail()
filtro6
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV (lo crea :v)
filtro6.to_csv("Filtro6.csv")
```

```python
#FILTRO POR COMPARACIÓN
filtro7= df[df["PRECIO_UNITARIO"]>700]
filtro7
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV (lo crea :v)
filtro7.to_csv("Filtro7.csv")
```

```python
#FILTRO Y
filtro8= df[(df["PRECIO_UNITARIO"]<6000) & (df["FECHA_DOC"]=="2022-03-14")]
filtro8
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV (lo crea :v)
filtro8.to_csv("Filtro8.csv")
```

```python
#FILTRO O
filtro9= df[(df["COSTO_UNITARIO"]<6000) | (df["FECHA_DOC"]=="2022-03-14")]
filtro9
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV (lo crea :v)
filtro9.to_csv("Filtro9.csv")
```

```python
#FILTRO NOT
filtro10=df[~(df["FECHA_DOC"]== "2023-03-14")]
filtro10
```
